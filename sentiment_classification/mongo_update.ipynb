# Sentiment Classification - Data Preprocessing

```python
# Dependencies
import numpy as np
import pandas as pd
```

## Combine Datasets

```python
# Read in initial dataset
df1 = pd.read_json('../data_formatting/data_formatted.json')
df1 = df1.set_index('url')
df1.head()
```

```python
# Read in human ratings
df2 = pd.read_csv('../data_formatting/urls_human_sentiment.csv')
df2 = df2.set_index('url')['sentiment_human']
df2.head()
```

```python
# Combine dataframes
rated_df = pd.DataFrame.join(df1, df2, how='inner')
rated_df = rated_df.reset_index()
rated_df.head()
```

```python
rated_df
```

```python
sentiment_category = []
for x in rated_df['sentiment_human']:
    if x == -1:
        sentiment_category.append('negative')
    elif x == 0:
        sentiment_category.append('neutral')
    else: sentiment_category.append('positive')
```

```python
sentiment_category
```

```python
rated_df['sentiment_category'] = sentiment_category
```

```python
rated_df
```

```python
rated_df.to_json('updated_mongo_dataset.json', orient='records')
```

```python
test_df = pd.read_json('updated_mongo_dataset.json')
test_df
```

```python
test_df['sentiment_category'].value_counts()
```

```python
# Determine total number of each sentiment by domain
df2 = test_df.groupby(['source', 'sentiment_category']).count().reset_index()
df3 = df2[['source', 'sentiment_category', 'url']].rename(columns={'url': 'count'})
df3
```

```python
# Determine total number of each sentiment
df4 = test_df.groupby(['sentiment_category']).count().reset_index()
df5 = df4[['sentiment_category', 'url']].rename(columns={'url': 'count'})
df5
```

```python
# Format for JSON
all_domain = []
for index, row in df5.iterrows():
    all_domain.append({'aggregation': 'all', 'sentiment_category': row['sentiment_category'], 'count': row['count']})
all_domain
```

```python
# Format for JSON
by_domain = []
for index, row in df3.iterrows():
    by_domain.append({'aggregation': 'domain', 'sentiment_category': row['sentiment_category'], 'count': row['count'], 'source': row['source']})
by_domain
```

```python
import json
```

```python
# Save as JSON file - this was uploaded to mongodb
with open('sentiment_counts_all_updated.json', 'w') as outfile:
    json.dump(all_domain, outfile)
```

```python
# Save as JSON file - this was uploaded to mongodb
with open('sentiment_counts_domains_updated.json', 'w') as outfile:
    json.dump(by_domain, outfile)
```
