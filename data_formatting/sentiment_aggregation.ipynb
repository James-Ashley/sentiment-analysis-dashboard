```python
# Dependencies
import pandas as pd
import json
```

```python
# Read in data
df = pd.read_json('data_formatted.json')
df
```

```python
# Add sentiment category
df['sentiment_category'] = 'neutral'
df.loc[df['compound_score'] > 0.2, 'sentiment_category'] = 'positive'
df.loc[df['compound_score'] < -0.2, 'sentiment_category'] = 'negative'
df
```

```python
# Determine total number of each sentiment by domain
df2 = df.groupby(['source', 'sentiment_category']).count().reset_index()
df3 = df2[['source', 'sentiment_category', 'url']].rename(columns={'url': 'count'})
df3
```

```python
# Determine total number of each sentiment
df4 = df.groupby(['sentiment_category']).count().reset_index()
df5 = df4[['sentiment_category', 'url']].rename(columns={'url': 'count'})
df5
```

```python
# Format for JSON
all_domain = []
for index, row in df5.iterrows():
    all_domain.append({'aggregation': 'all', 'sentiment_category': row['sentiment_category'], 'count': row['count']})
all_domain
```

```python
# Format for JSON
by_domain = []
for index, row in df3.iterrows():
    by_domain.append({'aggregation': 'domain', 'sentiment_category': row['sentiment_category'], 'count': row['count'], 'source': row['source']})
by_domain
```

```python
# Save as JSON file - this was uploaded to mongodb
with open('sentiment_counts_all.json', 'w') as outfile:
    json.dump(all_domain, outfile)
```

```python
# Save as JSON file - this was uploaded to mongodb
with open('sentiment_counts_domains.json', 'w') as outfile:
    json.dump(by_domain, outfile)
```
