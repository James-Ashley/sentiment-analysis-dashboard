```python
# Dependencies
import pandas as pd
```

```python
# Import in complete dataset
df = pd.read_json('final_data_cleaned.json')
df
```

```python
# Rename columns for easier use in Flask API
df = df.rename(columns={'Keyword': 'keyword', 
                        'Source': 'source', 
                        'Author': 'author', 
                        'Title': 'title', 
                        'URL': 'url', 
                        'Published': 'published', 
                        'compound score': 'compound_score', 
                        'negative score': 'negative_score',
                        'positive score': 'positive_score',
                        'neutral score': 'neutral_score', 
                        'Text': 'text_excerpt', 
                        'Full Text Clean': 'text_complete',})
```

```python
# Check column names
df.head()
```

```python
# Extract column names for documentation
df.columns
```

```python
# Reorder columns
df = df[['keyword', 'source', 'author', 'title', 'url',
       'published', 'compound_score', 'negative_score', 'positive_score',
       'neutral_score', 'text_excerpt', 'text_complete']]
df.head()
```

```python
# Save as JSON file
df.to_json('data_formatted.json', orient='records')
```
